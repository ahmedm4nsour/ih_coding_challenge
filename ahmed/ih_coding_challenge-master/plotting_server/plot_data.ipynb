```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the data
data = pd.read_csv("../data_processing/plot_data.csv")

# Plot
plt.plot(data['Position'], data['Original_Intensity'], label='Original')
plt.plot(data['Position'], data['Filtered_Intensity'], label='Filtered')
plt.xlabel('Position')
plt.ylabel('Intensity')
plt.title('Original and Filtered Data')
plt.legend()
plt.show()
```
